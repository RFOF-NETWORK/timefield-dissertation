# Simulation: Gravitationsgradienten und lokale Zeitdilatation

Dieses Notebook simuliert die relativistischen Zeitgradienten, die durch
Gravitationspotentiale im planetaren und solaren Maßstab entstehen.

Die Ergebnisse fließen in die relativistische Schicht des integrierten
Zeitfeldmodells ein.

```python
import numpy as np
import matplotlib.pyplot as plt

# Physikalische Konstanten
G = 6.67430e-11       # Gravitationskonstante (SI)
c = 299792458         # Lichtgeschwindigkeit (m/s)

# Parameter für Erde und Sonne
M_earth = 5.972e24    # kg
R_earth = 6.371e6     # m

M_sun = 1.989e30      # kg
R_sun = 6.9634e8      # m

def schwarzschild_time_dilation(M, r):
    return np.sqrt(1 - 2*G*M/(r*c**2))

r_earth = np.linspace(R_earth, R_earth + 1e6, 2000)
r_sun = np.linspace(R_sun, R_sun + 5e8, 2000)

tau_earth = schwarzschild_time_dilation(M_earth, r_earth)
tau_sun = schwarzschild_time_dilation(M_sun, r_sun)
```

## Zeitgradienten

Die Zeitgradienten ergeben sich aus der Ableitung der Zeitdilatationsfunktion:

$$ \nabla T(r) = \frac{d}{dr} \left( \sqrt{1 - \frac{2GM}{rc^2}} \right) $$

```python
grad_earth = np.gradient(tau_earth, r_earth)
grad_sun = np.gradient(tau_sun, r_sun)
```

## Plot: Zeitdilatation und Zeitgradienten

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(r_earth - R_earth, tau_earth, label='Erde')
plt.plot(r_sun - R_sun, tau_sun, label='Sonne')
plt.title('Zeitdilatation im Gravitationsfeld')
plt.xlabel('Höhe über Oberfläche (m)')
plt.ylabel('dτ/dt')
plt.legend()

plt.subplot(1,2,2)
plt.plot(r_earth - R_earth, grad_earth, label='Erde')
plt.plot(r_sun - R_sun, grad_sun, label='Sonne')
plt.title('Zeitgradienten')
plt.xlabel('Höhe über Oberfläche (m)')
plt.ylabel('∂(dτ/dt)/∂r')
plt.legend()

plt.tight_layout()
plt.show()
```

## Ergebnis

Dieses Notebook zeigt:

- die relativistische Zeitdilatation im Gravitationsfeld von Erde und Sonne,
- die daraus resultierenden Zeitgradienten,
- die Relevanz dieser Gradienten für die relativistische Schicht des Zeitfeldes.

Die resultierenden Gradienten fließen direkt in die Komponenten:

$$ \mathcal{T}^{(r)}_{00} = g_{00} $$
$$ \nabla T^{(r)} = \partial_r \sqrt{g_{00}} $$

und beeinflussen die Interferenz mit planetaren, solaren und galaktischen Feldern.
