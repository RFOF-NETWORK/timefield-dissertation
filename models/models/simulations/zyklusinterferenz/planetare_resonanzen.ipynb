# Simulation: Planetare Resonanzen und Interferenzstrukturen

Dieses Notebook simuliert die Interferenz planetarer Zyklen und die daraus
entstehenden Resonanzmuster. Diese Strukturen bilden die Grundlage der
    planetaren Schicht im integrierten Zeitfeldmodell.

```python
import numpy as np
import matplotlib.pyplot as plt

# Beispielhafte planetare Frequenzen (in 1/Jahr)
omega_earth = 2 * np.pi / 1.0       # Erdjahr
omega_jupiter = 2 * np.pi / 11.86   # Jupiterjahr
omega_saturn = 2 * np.pi / 29.46    # Saturnjahr

# Zeitachse
t = np.linspace(0, 200, 50000)  # 200 Jahre
```

## Planetare Signale

Die planetaren Zyklen werden als harmonische Signale modelliert:

```python
A_e = 1.0
A_j = 0.7
A_s = 0.5

earth = A_e * np.cos(omega_earth * t)
jupiter = A_j * np.cos(omega_jupiter * t)
saturn = A_s * np.cos(omega_saturn * t)
```

## Interferenz: Jupiter–Saturn Resonanz

Die synodische Frequenz ergibt sich aus:

$$ \omega_{JS} = |\omega_J - \omega_S| $$

```python
omega_js = abs(omega_jupiter - omega_saturn)
resonance_js = np.cos(omega_js * t)
```

## Gesamtsignal und Fourier-Spektrum

```python
signal = earth + jupiter + saturn

# Fourier-Transformation
freqs = np.fft.rfftfreq(len(t), d=(t[1] - t[0]))
spectrum = np.abs(np.fft.rfft(signal))
```

## Plot: Interferenz und Spektrum

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(t, signal, linewidth=0.5)
plt.title('Interferenz planetarer Zyklen')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Amplitude')

plt.subplot(1,2,2)
plt.plot(freqs, spectrum)
plt.title('Fourier-Spektrum')
plt.xlabel('Frequenz (1/Jahr)')
plt.ylabel('Spektrale Stärke')

plt.tight_layout()
plt.show()
```

## Ergebnis

Dieses Notebook zeigt:

- die planetaren Grundzyklen von Erde, Jupiter und Saturn,
- die synodische Resonanz Jupiter–Saturn,
- die Interferenzstruktur über 200 Jahre,
- das Fourier-Spektrum mit klaren Resonanzpeaks.

Diese Resonanzen fließen direkt in die planetare Schicht des Zeitfeldtensors ein:

$$ \mathcal{T}^{(p)}_{ij}(t) = \sum_k A_k \cos(\omega_k t + \phi_k) $$

und beeinflussen die Interferenz mit solaren und galaktischen Feldern.
