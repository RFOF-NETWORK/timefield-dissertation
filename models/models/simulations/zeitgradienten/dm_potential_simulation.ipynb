# Simulation: Dunkle-Materie-Potential und Zeitgradienten

Dieses Notebook simuliert die Zeitgradienten, die durch ein galaktisches
Dunkle-Materie-Halo (NFW-Profil) erzeugt werden. Die resultierenden
Potentiale fließen in die galaktische Schicht des integrierten Zeitfeldmodells ein.

```python
import numpy as np
import matplotlib.pyplot as plt

# NFW-Parameter
rho0 = 0.3      # zentrale Dichte (arbitrary units)
rs = 20.0       # Skalenradius (kpc)
G = 4.302e-6    # Gravitationskonstante (kpc * (km/s)^2 / Msun)

def nfw_potential(r):
    x = r / rs
    return -4 * np.pi * G * rho0 * rs**2 * np.log(1 + x) / x

r = np.linspace(0.1, 200, 2000)
phi = nfw_potential(r)
```

## Zeitgradienten aus dem Potential

Die Zeitgradienten ergeben sich aus der Ableitung des Potentials:

$$ \nabla T \propto \partial_r \Phi(r) $$

```python
dphi_dr = np.gradient(phi, r)
time_gradient = -dphi_dr  # Vorzeichenkonvention
```

## Plot: Potential und Zeitgradient

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(r, phi)
plt.title('NFW-Potential')
plt.xlabel('Radius r (kpc)')
plt.ylabel('Φ(r)')

plt.subplot(1,2,2)
plt.plot(r, time_gradient)
plt.title('Zeitgradient dΦ/dr')
plt.xlabel('Radius r (kpc)')
plt.ylabel('dΦ/dr')

plt.tight_layout()
plt.show()
```

## Ergebnis

Dieses Notebook zeigt:

- das NFW-Potential eines Dunkle-Materie-Halos,
- die daraus resultierenden Zeitgradienten,
- die Relevanz dieser Gradienten für die galaktische Schicht des Zeitfeldes.

Die resultierenden Gradienten fließen direkt in die Komponente:

$$ \mathcal{T}^{(g)}_{00}(r) = \Phi_{\text{NFW}}(r) $$

und beeinflussen die Interferenz mit solaren und planetaren Zeitfeldern.
