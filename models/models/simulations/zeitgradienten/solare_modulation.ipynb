# Simulation: Solare Modulation und Interferenz mit planetaren Zyklen

Dieses Notebook simuliert die solaren Aktivitätszyklen (11‑Jahres‑Zyklus,
22‑Jahres‑Hale‑Zyklus, Gleissberg‑Modulation) und ihre Interferenz mit
planetaren Frequenzen. Diese Strukturen bilden die solare Schicht des
    integrierten Zeitfeldmodells.

```python
import numpy as np
import matplotlib.pyplot as plt

# Zeitachse: 500 Jahre
t = np.linspace(0, 500, 200000)
```

## Solare Grundzyklen

### 11‑Jahres‑Zyklus (Schwabe)
### 22‑Jahres‑Zyklus (Hale)
### Gleissberg‑Modulation (~80–100 Jahre)

```python
omega_11 = 2 * np.pi / 11
omega_22 = 2 * np.pi / 22
omega_g  = 2 * np.pi / 90

A_11 = 1.0
A_22 = 0.6
A_g  = 0.3

cycle_11 = A_11 * np.cos(omega_11 * t)
cycle_22 = A_22 * np.cos(omega_22 * t)
cycle_g  = A_g  * np.cos(omega_g  * t)
```

## Gesamtsignal der solaren Aktivität

Die solare Aktivität ergibt sich als Überlagerung aller drei Zyklen.

```python
solar_signal = cycle_11 + cycle_22 + cycle_g
```

## Kopplung mit planetaren Frequenzen

Beispiel: Jupiter–Saturn‑Resonanz als externer Modulator.

```python
omega_j = 2 * np.pi / 11.86
omega_s = 2 * np.pi / 29.46
omega_js = abs(omega_j - omega_s)

    # Modulator
mod_js = np.cos(omega_js * t)

    # Interferenz solare Aktivität × planetare Resonanz
interference = solar_signal * (1 + 0.2 * mod_js)
```

## Fourier-Spektrum der solaren Interferenz

```python
freqs = np.fft.rfftfreq(len(t), d=(t[1] - t[0]))
spectrum = np.abs(np.fft.rfft(interference))
```

## Plot: Solare Aktivität, Modulation und Spektrum

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(t, interference, linewidth=0.4)
plt.title('Solare Aktivität × Jupiter–Saturn‑Modulation')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Amplitude')

plt.subplot(1,2,2)
plt.plot(freqs, spectrum)
plt.title('Fourier-Spektrum der solaren Interferenz')
plt.xlabel('Frequenz (1/Jahr)')
plt.ylabel('Spektrale Stärke')

plt.tight_layout()
plt.show()
```

## Ergebnis

Dieses Notebook zeigt:

- die solaren Grundzyklen (11, 22, 90 Jahre),
- die Interferenz mit der Jupiter–Saturn‑Resonanz,
- die resultierende modulierte solare Aktivität,
- das Fourier-Spektrum mit kombinierten Peaks.

Diese Struktur fließt direkt in die solare Schicht des Zeitfeldtensors ein:

$$ \mathcal{T}^{(s)}_{0i}(t) = A_{\odot}(t) B_{\odot,i}(t) $$

und beeinflusst die Interferenz mit planetaren und galaktischen Feldern.
