# Analyse: Interferenz der Zeitfeld-Zyklen

Dieses Notebook untersucht die Interferenzstrukturen zwischen den verschiedenen
Zyklen des integrierten Zeitfeldmodells:

- planetare Frequenzen
- solare Aktivitätszyklen
- galaktische Modulationen
- relativistische Gradienten

Ziel ist es, Schwebungen, Resonanzen und Langzeitmodulationen sichtbar zu machen.

## 1. Import der Bibliotheken

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

## 2. Beispielhafte Zyklusdaten

Die Frequenzen werden exemplarisch definiert, um Interferenzstrukturen zu simulieren.

```python
# Beispielhafte Frequenzen (in 1/Jahr)
omega_planet = 1/11.86   # Jupiter
omega_solar = 1/11       # Schwabe-Zyklus
omega_galactic = 1/240e6 # galaktische Modulation (vereinfacht)

# Zeitachse
t = np.linspace(0, 500, 200000)  # 500 Jahre
```

## 3. Signale der einzelnen Schichten

```python
A_p = 1.0
A_s = 0.6
A_g = 0.2

planet = A_p * np.cos(2*np.pi*omega_planet * t)
solar = A_s * np.cos(2*np.pi*omega_solar * t)
galactic = A_g * np.cos(2*np.pi*omega_galactic * t)
```

## 4. Interferenzsignal

```python
signal = planet + solar + galactic
```

## 5. Fourier-Spektrum

```python
freqs = rfftfreq(len(t), d=(t[1] - t[0]))
spectrum = np.abs(rfft(signal))
```

## 6. Plot: Interferenz und Spektrum

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(t[:20000], signal[:20000], linewidth=0.5)
plt.title('Interferenz der Zeitfeld-Zyklen')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Amplitude')

plt.subplot(1,2,2)
plt.plot(freqs, spectrum)
plt.title('Fourier-Spektrum')
plt.xlabel('Frequenz (1/Jahr)')
plt.ylabel('Spektrale Stärke')
plt.xlim(0, 0.2)

plt.tight_layout()
plt.show()
```

## 7. Interpretation

- Die planetare Frequenz erzeugt schnelle Oszillationen.
- Die solare Frequenz erzeugt mittelfristige Modulationen.
- Die galaktische Frequenz erzeugt extrem langsame Schwebungen.

Die Interferenz dieser Schichten bildet die Grundlage der beobachteten Zeitgradienten im integrierten Zeitfeldmodell.
