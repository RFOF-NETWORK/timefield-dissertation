# Analyse: Galaktische Zeitgradienten

Dieses Notebook untersucht die galaktischen Beiträge zur Zeitmodulation.
Im integrierten Zeitfeldmodell beschreibt die galaktische Schicht (g)
langsame, großskalige Veränderungen der Zeitrate, verursacht durch:

- Dunkle-Materie-Potentiale,
- Spiralarmstruktur,
- vertikale Oszillation der Sonne durch die galaktische Ebene,
- radiale Oszillation im galaktischen Gravitationsfeld.

## 1. Import der Bibliotheken

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

## 2. Modellierung galaktischer Oszillationen

Die Sonne führt zwei relevante Bewegungen aus:

1. **Vertikale Oszillation** durch die galaktische Scheibe (Periode ~ 60 Mio. Jahre)
2. **Radiale Oszillation** um den galaktischen Mittelpunkt (Periode ~ 240 Mio. Jahre)

Wir modellieren beide als harmonische Signale.

```python
# Frequenzen (in 1/Jahr)
omega_vert = 1 / (60e6)
omega_rad  = 1 / (240e6)

# Zeitachse: 500 Mio. Jahre
t = np.linspace(0, 500e6, 200000)
```

## 3. Galaktische Potentialmodulationen

Die galaktischen Zeitgradienten werden als Kombination der beiden Oszillationen modelliert:

```python
A_vert = 1.0
A_rad  = 0.6

vert = A_vert * np.cos(2*np.pi*omega_vert * t)
rad  = A_rad  * np.cos(2*np.pi*omega_rad  * t)

galactic_signal = vert + rad
```

## 4. Zeitgradientenfunktion

Die galaktische Schicht beeinflusst die Zeitrate über:

$$ T^{(g)}(t) = A_v \cos(\omega_v t) + A_r \cos(\omega_r t) $$

Die resultierende Zeitrate ist proportional zu:

$$ \frac{d\tau}{dt} \propto 1 + T^{(g)}(t) $$

```python
time_rate = 1 + galactic_signal
```

## 5. Visualisierung der galaktischen Zeitgradienten

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(t, galactic_signal)
plt.title('Galaktische Oszillationen')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Amplitude')

plt.subplot(1,2,2)
plt.plot(t, time_rate)
plt.title('Galaktische Zeitgradienten')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('dτ/dt (normiert)')

plt.tight_layout()
plt.show()
```

## 6. Fourier-Spektrum der galaktischen Oszillationen

Zur Identifikation dominanter galaktischer Frequenzen wird das Spektrum berechnet.

```python
freqs = np.fft.rfftfreq(len(t), d=(t[1] - t[0]))
spectrum = np.abs(np.fft.rfft(galactic_signal))
```

## 7. Dominante Frequenzen (Peaks) im Spektrum

```python
peaks, _ = find_peaks(spectrum, height=np.max(spectrum)*0.1)
dominant_freqs = freqs[peaks]
dominant_freqs[:10]
```

## 8. Plot: Fourier-Spektrum mit Peaks

```python
plt.figure(figsize=(10,6))
plt.plot(freqs, spectrum)
plt.scatter(dominant_freqs, spectrum[peaks], color='red')
plt.title('Fourier-Spektrum der galaktischen Oszillationen')
plt.xlabel('Frequenz (1/Jahr)')
plt.ylabel('Spektrale Stärke')
plt.xlim(0, 1e-6)
plt.show()
```

## 9. Galaktischer Tensorbeitrag (Proxy)

Im vollen Zeitfeldmodell würde der galaktische Beitrag als Tensor
$T^{(g)}_{\mu\nu}$ auftreten. Hier verwenden wir einen skalarisierten Proxy:

```python
T_g = galactic_signal  # skalarer Proxy für T^{(g)}_{\mu\nu}
T_g[:10]
```

## 10. Beispielhafte Kopplung mit solarem Zyklus

Zur Illustration der Interferenz kann die galaktische Schicht mit einem solaren Zyklus gekoppelt werden.

```python
omega_solar = 1/11  # Schwabe-Zyklus
solar = 0.5 * np.cos(2*np.pi*omega_solar * t)

interferenz_gs = galactic_signal + solar
```

## 11. Plot: Interferenz galaktisch × solar

```python
plt.figure(figsize=(12,5))
plt.plot(t[:20000], interferenz_gs[:20000], linewidth=0.5)
plt.title('Interferenz: galaktische Oszillation × solarer Zyklus')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Amplitude')
plt.show()
```

## 12. Interpretation

- Die galaktischen Oszillationen erzeugen extrem langsame Modulationen.
- Im Spektrum erscheinen sehr niederfrequente Peaks.
- Die Kopplung mit solaren Zyklen erzeugt komplexe Interferenzstrukturen.

Diese Strukturen bilden die Grundlage der galaktischen Schicht im integrierten Zeitfeldmodell.
